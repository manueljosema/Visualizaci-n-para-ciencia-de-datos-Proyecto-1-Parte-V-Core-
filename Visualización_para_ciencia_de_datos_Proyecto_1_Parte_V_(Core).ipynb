```python
# leer el documento csv
data = pd.read_csv('/content/retail_sales_dataset.csv')
```

```python

# Calculate the correlation matrix for numerical columns
correlation_matrix = data[['Quantity', 'Price per Unit', 'Total Amount']].corr()

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True, cbar_kws={'shrink': 0.8})
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Configura el tamaño y la estructura de los subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Análisis Exploratorio de Ventas Minoristas", fontsize=16)

# Gráfico 1: Relación entre Edad y Cantidad
sns.scatterplot(data=data, x='Age', y='Quantity', ax=axs[0, 0], color="skyblue")
axs[0, 0].set_title("Relación entre Edad y Cantidad")
axs[0, 0].grid(True)  # Añadir cuadrícula
axs[0, 0].annotate("Edad típica de compra alta", xy=(30, 3), xytext=(45, 4),
                   arrowprops=dict(facecolor='black', shrink=0.05))

# Gráfico 2: Distribución de Total Amount por Categoría de Producto
sns.boxplot(data=data, x='Product Category', y='Total Amount', ax=axs[0, 1], hue="Product Category", dodge=False)
axs[0, 1].set_title("Distribución del Monto Total por Categoría de Producto")
axs[0, 1].grid(True)
axs[0, 1].legend([], [], frameon=False)  # Ocultar leyenda duplicada

# Gráfico 3: Total Amount vs. Quantity
sns.scatterplot(data=data, x='Quantity', y='Total Amount', hue='Product Category', ax=axs[1, 0])
axs[1, 0].set_title("Total Amount vs. Quantity por Categoría")
axs[1, 0].legend(title='Categoría')
axs[1, 0].grid(True)

# Gráfico 4: Cantidad de compras por género
sns.countplot(data=data, x='Gender', ax=axs[1, 1], hue="Gender", dodge=False)
axs[1, 1].set_title("Cantidad de Compras por Género")
axs[1, 1].grid(True)
axs[1, 1].legend([], [], frameon=False)  # Ocultar leyenda duplicada

# Ajuste de diseño
plt.tight_layout(pad=2.0, h_pad=2.5, w_pad=3.0)
plt.subplots_adjust(top=0.9)  # Ajusta el espacio para el título principal
plt.show()
```
